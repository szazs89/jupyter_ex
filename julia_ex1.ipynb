## Benchmarking

We calculate the floating point operations per second from the elapsed time of multiplying two matrices __A__ and __B__.

```julia
n = Int(8192/2);
aa = rand(Float32,n,n);
bb = rand(Float32,n,n);
```

```julia
etime = @elapsed aa*bb
```

```julia
GFlops = 2*n^3/(etime*1e+9)
```

```julia
?@elapsed
```

## Plotting

```julia
n = 200;
t = 0:10/n:10;
y = [ sin(ti) for ti in t ];
```

```julia
using FFTW
```

```julia
Y = fft(y)
Y = [ abs(Y[i]) for i in 1:Int(n/2)+1 ]
f = 0:1/t[n+1]:Int(1/t[2]/2)
```

```julia
using PyPlot
```

```julia
figure(1,figsize=(10,4))
subplot(121)
plot(t,y,c="red",lw=1)
title("sin(t)")
grid()
xlabel("t[s]")

subplot(122)
plot(f,Y)
title("FFT")
grid()
xlabel("n[Hz]")
```
